```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import pearsonr, spearmanr
%matplotlib inline
```

```python
df = pd.read_csv('alpha24_step15Stud.csv')
df.head()
```

```python
df['temp'] = df['temp'].apply(lambda x: x.replace(',','.')).astype('float64')
df['Ts'] = df['Ts'].apply(lambda x: x.replace(',','.')).astype('float64')
```

```python
df.head()
```

```python
plt.scatter(df['temp'],df['Ts'])
```

```python
pearsonr(df['temp'],df['Ts'])
```

temperatura chłodziarki i temperatura otoczenia są silnie skorelowane

```python
spearmanr(df['temp'],df['Ts'])
```

```python
x = []
corr = []
for period in range(50,55):
    df_diff = df.diff(periods=period)[period:]
    for i in range(100, 1 , -1):
        x.append(i)
        corr.append(pearsonr(df_diff['temp'][i:],df_diff['Ts'][:-i])[0])

    for i in range(1, 200):
        x.append(-i)
        corr.append(pearsonr(df_diff['temp'][:-i],df_diff['Ts'][i:])[0])

    plt.plot(x, corr)
```

```python
max(corr)
```

```python
optimal_shift = abs(x[corr.index(max(corr))])
x[corr.index(max(corr))]
```

Największa autokorelacja występuje dla  temperatury z przed 17 minut

```python
plt.scatter(df['temp'][:-optimal_shift],df['Ts'][optimal_shift:])
```
